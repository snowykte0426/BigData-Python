```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('./남북한발전전력량.xlsx', engine='openpyxl')

df_ns = df.iloc[[0, 5], 2:]

df_ns.index = ['South', 'North']

df_ns.columns = df_ns.columns.astype(int)

print(df_ns.head())
print('\n')

df_ns.plot()
```

```python
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot()
```

```python
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot(kind='bar')
```

```python
tdf_ns = df_ns.T.apply(pd.to_numeric, errors='coerce')

tdf_ns.plot(kind='hist')
```
